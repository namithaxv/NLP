```python
# BAG OF WORDS
```

```python
class Category:
  BOOKS = "BOOKS"
  CLOTHING = "CLOTHING"

train_x = ["i love the book", "this is a great book", "the fit is great", "i love the shoes"]
train_y = [Category.BOOKS, Category.BOOKS, Category.CLOTHING, Category.CLOTHING]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(binary=True)
train_x_vectors = vectorizer.fit_transform(train_x)

print(vectorizer.get_feature_names_out())
print(train_x_vectors.toarray())
```

```python
from sklearn import svm

clf_svm = svm.SVC(kernel='linear')
clf_svm.fit(train_x_vectors, train_y)
```

```python
test_x = vectorizer.transform(['i love the books'])

clf_svm.predict(test_x)
```

```python
# Sample data
corpus = [
    'Hindu communalist government because if',
    'say that if they do that then you',
    'immediately run into the idea of the',
    'temple that Modi government is being',
    'accused of being communal because they',
    'built the temple and because they seem',
    'to demolish all partitions between',
    'between the state between the government'
    'and religion the Hindu religion that is a',
    'reality but that is not how it will play'
    'out in elections because elections are',
    'not like a college debate these are',
    'elections people going with their',
    'emotions they go to vote'
    'so what does the what does the',
    'opposition do the three opposition',
    'parties we have seen all constituents of',
    'the India block they found some clarity',
    'they left for example that is easy to',
    'understand Pinar Vian has spoken out and',
    'he said that the church and the state',
    'should always be separate my words not',
    'his he said that in our secular system',
    'the state and the and the religion',
    'should be kept very clinically apart'
]

# Create an instance of CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the corpus to create the Bag of Words representation
X = vectorizer.fit_transform(corpus)

# Get the feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert the sparse matrix to a dense array for better visibility
X_dense = X.toarray()

# Create a DataFrame to display the results
import pandas as pd

df = pd.DataFrame(X_dense, columns=feature_names)
print(df)
```

```python
import spacy

# Load the spaCy NLP model (you may need to install it first: pip install spacy)
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "I love dogs and cats in New York at the Samsung store "

# Process the text with spaCy NLP model
doc = nlp(text)

# Extract named entities
entities = [(ent.text, ent.label_) for ent in doc.ents]

# Print the named entities
for entity in entities:
    print(f"Entity: {entity[0]}, Type: {entity[1]}")
```
